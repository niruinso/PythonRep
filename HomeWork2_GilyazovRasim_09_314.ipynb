```python
class Invoice:
  def __init__(self, invoiceNumber, amount):
    self.__invoiceNumber = invoiceNumber
    self.__amount = amount

  @property
  def invoiceNumber(self):
    return self.__invoiceNumber

  @property
  def amount(self):
    return self.__amount

  @invoiceNumber.setter
  def invoiceNumber(self, new_invoiceNumber):
    if new_invoiceNumber < 0:
      print("Bad choice")
    else:
      self.__invoiceNumber = new_invoiceNumber

  @amount.setter
  def amount(self, new_amount):
    if new_amount < 0:
      print("Bad choice!")
    else:
      self.__amount = new_amount;

x = Invoice(100, 15)
print(x.amount)
x.amount = -5
print(x.amount)
```

```python
class University:
  class Faculty:
    def __init__(self, name, students):
      self.name = name
      self.students = students[:]

    def add_student(self, student):
      self.students.append(student)

  def __init__(self, faculties):
    self.faculties = faculties[:]

  def add_faculty(self, faculty):
    self.faculties.append(faculty)

  def show(self):
    for elem in self.faculties:
      print(f'faculty name: {elem.name}; students: {elem.students[:]}')

field1 = ['rasim', 'ramzan', 'artem']
field2 = ['123', '234', '345']

x = University([University.Faculty('pmi', field1)])
x.show()
print('Adding new faculty: "pm"')
x.add_faculty(University.Faculty('pm', field2))
x.show()
```

```python
class Subject:
  def __init__(self, name, rating):
    self._name = name
    self._rating = rating


class Student(Subject):
  def __init__(self, subjects):
    self.subjects = subjects[:]

  def show(self):
    for elem in self.subjects:
      print(f'Subject: {elem._name}, rating: {elem._rating}')


Rasim = Student([Subject("Math", 5), Subject("Physics", 4), Subject("Chemistry", 5)])

Rasim.show()
```

```python
import abc

class AudioDevice:
  @abc.abstractmethod
  def playSound(self): pass


class Headphones(AudioDevice):
  def __init__(self, sound):
    self.sound = sound

  def playSound(self):
    print(f'In headphones you can hear: {self.sound}')


class Speakers(AudioDevice):
  def __init__(self, sound):
    self.sound = sound

  def playSound(self):
    print(f'Speaker says you: {self.sound}')


headphone = Headphones("Hello")
speaker = Speakers("World")

headphone.playSound()
speaker.playSound()
```

```python
from abc import ABC, abstractmethod

class IPayment(ABC):
  @abstractmethod
  def updateAmount(self, amount: float): pass

  @abstractmethod
  def getPaymentDetails(self): pass


class Payment(IPayment):
  def __init__(self, amount):
    self.amount = amount

  def updateAmount(self, amount: float):
    self.amount = amount

  def getPaymentDetails(self):
    print(f'Current amount to pay: {self.amount}')


someClass = Payment(500)

someClass.getPaymentDetails()
someClass.updateAmount(300)
someClass.getPaymentDetails()
```

```python
class Invoice:
  def __init__(self, amount):
    self.amount = amount

  def sendInvoice(self):
    print(f'Current amount to pay: {self.amount}')

class EmailInvoice(Invoice):
  def sendInvoice(self):
    print(f'Current amount to pay by Email: {self.amount}')


class PaperInvoice(Invoice):
  def sendInvoice(self):
    print(f'Current amount to pay by Paper: {self.amount}')


email = EmailInvoice(123)
paper = PaperInvoice(256)

email.sendInvoice()
paper.sendInvoice()
```
