```python
N = int(input())

for num1 in range(2, N):
  count1 = 1
  for i in range(2, round((num1 ** 0.5)) + 1) :
    if (num1 % i == 0 and i * i != num1):
      count1 += i
      count1 += num1 / i
    elif num1 % i == 0:
      count1 += i
  num2 = count1
  count2 = 1
  for i in range(2, round((num2 ** 0.5)) + 1) :
    if (num2 % i == 0 and i * i != num2):
      count2 += i
      count2 += num2 / i
    elif num2 % i == 0:
      count2 += i
  if(count2 == num1):
    print(int(num1), int(num2), sep = ' ')
```

```python
def Simm(S, i, j):
  if j <= i:
    return 1
  if(S[i] == S[j]):
    return 1 * Simm(S, i + 1, j - 1)
  else:
    return 0

s = "abcdeffedcba"
print(bool(Simm(s, 0, len(s) - 1)))
```

```python
from math import factorial

eps = 0.001
prev_sum = 0
this_sum = factorial(1) / factorial(2 * 1)
i = 2
while(this_sum - prev_sum >= eps) :
  prev_sum = this_sum
  this_sum += factorial(i) / factorial(2 * i)
  i += 1
print(this_sum)
```
