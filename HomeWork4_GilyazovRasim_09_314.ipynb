```python
from random import shuffle
import locale
import datetime

def start(file_name, txt):
  with open(file_name, "w", encoding='utf-8') as file:
    file.write(txt)

def read_file(file_name):
    file = open(file_name, "r", encoding='utf-8')
    content = file.read().split('\n')
    file.close()
    sz = len(content)
    my_arr = [[] for _ in range(sz)]
    for i in range(sz):
        my_arr[i] = content[i].split('|')
    return sz, my_arr

def testing(my_arr, arr_of_indexes, sz):
    correct_answers_counter = 0
    for i in range(sz):
        print(f"Вопрос номер {i + 1}: {my_arr[i][0]}")
        shuffle(arr_of_indexes)
        print("Варианты ответов:")
        for j in range(len(arr_of_indexes)):
            print(f"{j+1}) {my_arr[i][arr_of_indexes[j]]}")

        answer = -1
        while(answer == -1):
            answer = get_answer()
            if(answer == -1):
                print("Некорректный ввод! Попробуйте еще раз: требуется ввести число от 1 до 5")

        if my_arr[i][arr_of_indexes[answer - 1]] == my_arr[i][6]:
            print("Правильно!")
            correct_answers_counter +=1
        else:
            print("К сожалению, вы ошиблись!")
        if(i != sz - 1):
            print("Переходим к следующему вопросу...\n\n")
        else:
            print("\n\nТестирование завершено!")
    return correct_answers_counter

def write_out_result(file_name, begin, end, sz, counter):
    with open(file_name, 'w', encoding='utf-8') as file:
        file.write(f"Время начала теста: {begin}\n")
        file.write(f"Время окончания теста: {end}\n")
        file.write(f"Общее количество вопросов: {sz}\n")
        file.write(f"Количество правильных ответов: {counter}\n")
        file.write(f"Процент правильных ответов: {round(counter / sz * 100, 2)}%\n")

def print_result(sz, counter):
    print(f"Общее количество вопросов: {sz}")
    print(f"Количество правильных ответов: {counter}")
    print(f"Процент правильных ответов: {round(counter / sz * 100, 2)}%")

def get_answer():
    answer = 0
    try:
        answer = int(input("Ваш ответ: "))
    except ValueError:
        return -1
    return answer if answer >= 1 and answer <= 5 else -1

if __name__ == "__main__":
    txt = """Какой язык программирования используется для веб-разработки?|Python|JavaScript|Java|C++|Ruby|JavaScript
Столица Франции?|Париж|Москва|Вашингтон|Рим|Ницца|Париж
Какая планета в нашей Солнечной системе наиболее близка к Солнцу?|Марс|Венера|Земля|Меркурий|Юпитер|Меркурий
Какой газ составляет большую часть атмосферы Земли?|Кислород|Углекислый газ|Водород|Гелий|Азот|Азот
Кто написал “Ромео и Джульетту”?|Лев Толстой|Чарльз Диккенс|Уильям Шекспир|Джейн Остин|Марк Твен|Уильям Шекспир"""
    start("Questions.txt", txt)
    sz, my_arr = read_file("Questions.txt")
    arr_of_indexes = [i for i in range(1, 6)]

    begin = datetime.datetime.now().strftime("%Y-%m-%d1 %H:%M:%S")
    correct_answers_counter = testing(my_arr, arr_of_indexes, sz)
    end = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")

    write_out_result("Result.txt", begin, end, sz, correct_answers_counter)
    print_result(sz, correct_answers_counter)
```
