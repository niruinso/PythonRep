```python
rows = int(input("Enter the count of rows: "))
cols = int(input("Enter the count of cols: "))

fib_nums = [1, 1]
ind = 1

while fib_nums[ind] < max(rows, cols):
  fib_nums.append(fib_nums[ind] + fib_nums[ind - 1])
  ind += 1

matrix = [[0 for _ in range(cols)] for _ in range(rows)]

for row in range(rows):

  for col in range(cols):
    matrix[row][col] = int(input(f'arr[{row}][{col}] = '))
res = 0

for row in range(rows):
  if row in fib_nums:
    for col in range(cols):
      res += matrix[row][col] if col in fib_nums else 0

print(res)
```

```python
def foo(arr : dict) :
  sum = 0
  pr = 1
  for value in arr.values():
    sum += value
    pr *= value

  lst1 = []
  lst2 = []

  for k, v in arr.items():
    if(v % 2 == 0):
      lst1.append(k)
    elif(v % 2 == 1):
      lst2.append(k)

  return sum, pr, lst1, lst2

arr = {1:2, 2:3, 4:4, 5:1, 9:11}

print(foo(arr))
```

```python
def foo(arr : dict) :
  res1 = list(arr.keys())[0]
  res2 = list(arr.keys())[0]

  for k in arr.keys():
    if len(res1) < len(k):
      res1 = k
    if len(res2) > len(k):
      res2 = k

  return res1, res2

arr = {"some" : 1, "dict" : 2, "repeat" : 1}

print(foo(arr))
```

```python
def foo(arr : dict):
  sum = 0
  count = 0
  for value in arr.values():
    if isinstance(value, int):
      sum += value
      count += 1

  return sum / count

arr = {1:2, 2:3, 3:3.4, 4:"word"}

print(foo(arr))
```
